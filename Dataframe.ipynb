```python
import numpy as np
import pandas as pd
```

```python
# nous allons apprendre un csv (valeurs séparées par une virgule ou comma separated value)
# nous commencerons par les valeurs aléatoires, 101 dans seed est aléatoire mais permet d'avoir les memes partout
np.random.seed(101)
my_data = np.random.randint(0,101,(4,3))
my_data
# nous obtenons un tableau en deux dimensions avec des lignes et des colonnes
```

```python
# créons une liste d'index avec les noms des etats de USA qui contient quatre élements californie, New York,Arizona,Texas
my_index = ['CA','NY','AZ','TX']
```

```python
# on va aussi créer une liste des colonnes pour ma dataframe qui correspond aux mois de l'année
mycol = ['Jan','Feb','Mar']
```

```python
# créons un dataframe
# pour les dataframe , nous avons trois paramètres alors que pour les séries , nous avions que deux paramètres
df = pd.DataFrame(data= my_data,index=my_index,columns=mycol)
df
```

```python
# Je peux enlever les noms data,index et columns
df = pd.DataFrame(my_data,my_index,mycol)
df
```

```python
# nous pouvoir avoir des informations sur la dataset
df.info()
# nous avons les 4 entries ou index qui vont de CA à TX et la mémoire utilisée par le dataset qui est de 300 bytes
```
