# **opération groupby**

```python
df1 = pd.read_csv('/content/mpg.csv')
df1
```

```python
# on va voir comment les modèle de voiture ont évolue selon les années
# on regarde d'abord l'unicité des éléments de la colonne year
df1['model_year'].unique()
# on a choisi notre colonne catégorielle qui est model_year
```

```python
# utilisons groupby pour mesurer la moyenne des colonnes par catégorie model_year
df1.groupby('model_year').mean()
```

```python
# je peux trouver groupby d'une colonne
df1.groupby('model_year').mean()['mpg']
```

```python
# on peut aussi avoir un index multilayers ou multi niveaux
df1.groupby(['model_year','cylinders']).mean()
```

```python
# nous pouvons remplacer mean par describe et après transpose
df1.groupby('model_year').describe().transpose()
```

```python
# regardons maintenant ce qui se passe sur les index multi niveaux
year_cl = df1.groupby(['model_year','cylinders']).mean()
year_cl
```

```python
year_cl.index.levels
```

```python
# on peut obtenir une série ou la ligne pour model year et le nombre de cylindre
year_cl.loc[(70,8)]
# nous utilisons cette tuple, issue year_cl.index
```

```python
# techniquement ,nous pouvons avoir plus de deux niveaux
# ici l'index extérieur est model_year et l'interne est cylindres
# utilisons.xs pour faire du cross-section transversale
# imaginons que je veux tous les éléments du niveau model_year =70
year_cl.xs(key=70, level='model_year')
# je ne peux pas mettre deux ou plus Key,dans ce cas j'utilise .Loc[[70,80]]
```

```python
# imaginons que je vais avoir plusleurs instances du model_yer de 4 cylindres
year_cl.xs(key=4,level='cylinders')
```

```python
# trouvons les instances pour les cylindres 6 et 8
df1[df1['cylinders'].isin([6,8])].groupby(['model_year','cylinders']).mean()
```

## Fusion 'inner'(interne)

```python
registrations = pd.DataFrame({'reg_id':[1,2,3,4],'name':['Andrew','Bobo','Claire','David']})
logins = pd.DataFrame({'log_id':[1,2,3,4],'name':['Xavier','Andrew','Yolanda','Bobo']})
```

```python
pd.merge(registrations,logins,how='inner',on='name')
# cette opération ressemble à une intersection e deux ensembles, ce qu'on appelle le diagramme de venn
```

```python
pd.merge(left=registrations,right=logins,how='left',on='name')
# nous aimerons qu'il nous afficherons des éléments de la table de gauche et les éléments communs de eux tables
```

```python
pd.merge(left=registrations,right=logins,how='right',on='name')

# nous aimerons qu'il nous afficherons des éléments de la table de droite et les éléments communs de eux tables
```

```python
pd.merge(registrations,logins,how='outer',on='name')
# la fusion outer, afficher les éléments de deux tableaux et en évitant des doublons
# ici la colonne clé est la  colonne name
```

## fusion avec un index et nom de clés differents

```python
registrations = registrations.set_index('name')
registrations
```

```python
# ici nous avons l'index différent au datset logins, nous utiliserons les nouveaux paramètres
# ces paramètres sont: left_index,right_index,left_on,right_on
pd.merge(registrations,logins,left_index=True,right_on='name',how='inner')
```

```python
registrations = registrations.reset_index('name')
registrations
```

```python
registrations.columns=['reg_name','reg_id']
registrations
```

```python
# nous supposons que que nous avons deux colonnes clés différents de nos deux tables
pd.merge(registrations,logins,how='inner',left_on='reg_name',right_on='name')
```

```python
result = pd.merge(registrations,logins,how='inner',left_on='reg_name',right_on='name')
result
```

```python
# on peut supprimer la colonne doublon, en spécifiant axis=1 pour les colonnes
result.drop('reg_name',axis=1)
```
