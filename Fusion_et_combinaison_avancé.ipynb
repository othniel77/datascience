## Anti-join

```python
genres_tracks = genres.merge(top_tracks, on='movie_id', how='left', indicator=True)
print(genres_tracks.head())
```

```python
genres_tracks.info()
```

```python
gid_list = genres_tracks.loc[genres_tracks['_merge'] == 'left_only','movie_id']
print(gid_list.head())
```

```python
non_top_genres = genres[genres['movie_id'].isin(gid_list)]
print(non_top_genres.head())
```

```python
non_top_genres.info()
```

You performed an anti join by first merging the tables with a left join, selecting the ID of those employees who did not support a top customer, and then subsetting the original employee's table. From that, we can see that there are five employees not supporting top customers. Anti joins are a powerful tool to filter a main table

# Concatener ou combiner des dataframes ou des tables verticalements

```python
ridership = pd.read_pickle('/content/cta_ridership.p')
ridership.head()
```

```python
inv_jan = ridership[ridership['month'] == 1]
inv_jan.head()
```

```python
inv_feb = ridership[ridership['month'] == 2]
inv_feb.head()
```

```python
inv_mar = ridership[ridership['month'] == 3]
inv_mar.head()
```

```python
pd.concat([inv_jan, inv_feb, inv_mar])
```

```python
pd.concat([inv_jan, inv_feb, inv_mar],ignore_index=True)
```

```python
pd.concat([inv_jan, inv_feb, inv_mar],ignore_index=False,keys=['janvier','fevrier','mars'])
```

```python
pd.concat([inv_jan, inv_feb, inv_mar],ignore_index=False,keys=['janvier','fevrier','mars'])
```

```python
pd.concat([inv_jan, inv_feb],
sort=True)
```

```python
pd.concat([inv_jan, inv_feb],join='inner')
```
