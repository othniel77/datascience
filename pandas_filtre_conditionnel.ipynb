```python
# Nous allons extraire ou transformer ou castrer la colonne CC number
# en effet cette colonne de type entier et on peut le changer en colonne index
# je vais créer une fonction qui change ma colonne Num en string  et me retourne les 4 derniers valeurs
def last_four(num):
    return str(num)[-4:]
```

```python
# créons une fonction plus complexe
def yelp(price):
    if price < 10:
      return '$'
    elif price >= 10 and price < 30:
        return '$$'
    else:
        return '$$$'
```

```python
df['yelp']= df['total_bill'].apply(yelp)
df
```

```python
def quality (total_bill,tip):
    if tip/total_bill > 0.25 :
       return "generous"
    else:
       return "other"
```

```python
quality(19.90,1.01)
```

```python
df[['total_bill','tip']].apply(lambda df : quality(df['total_bill'],df['tip']),axis=1)
# ici on précise les deux colonnes et lambda df et qualité reprennent sa fonction .axis = 1 parce qu'on est sur les colonnes
```

```python
# il existe un moyen plus rapide de faire fonctionner cela
# par np.vectors
df['quality']=np.vectorize(quality)(df['total_bill'],df['tip'])
df.head()
```
